## Analyse World Happiness Record data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_2019 = pd.read_csv("2019.csv")

print("Summary Statistics:")
print(data_2019.describe())

top_countries = data_2019[['Country or region', 'Score']].head(5)
print("\nTop 5 Happiest Countries:")
print(top_countries)
```

```python
numeric_data = data_2019.drop(columns=['Country or region'])
plt.figure(figsize=(10, 6))
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Happiness Factors")
plt.show()
```

```python
data_2015 = pd.read_csv("2015.csv")
data_2016 = pd.read_csv("2016.csv")
data_2017 = pd.read_csv("2017.csv")
data_2018 = pd.read_csv("2018.csv")

# Add 'Year' to each dataframe before merging
data_2015['Year'] = 2015
data_2016['Year'] = 2016
data_2017['Year'] = 2017
data_2018['Year'] = 2018
data_2019['Year'] = 2019

data = pd.concat([data_2015, data_2016, data_2017, data_2018, data_2019], ignore_index=True)
```

```python
top_5_countries = data.groupby('Year').apply(lambda x: x.nsmallest(5, 'Overall rank')).reset_index(drop=True)

# Trend: Happiness Score
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Score', hue='Country or region', data=top_5_countries, marker='o')
plt.title('Happiness Score Trends for Top 5 Countries (2015-2019)')
plt.ylabel('Happiness Score')
plt.xlabel('Year')
plt.legend(title='Country')
plt.grid(True)
plt.show()

# Trend: Rank
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Overall rank', hue='Country or region', data=top_5_countries, marker='o')
plt.title('Rank Trends for Top 5 Countries (2015-2019)')
plt.ylabel('Overall Rank')
plt.xlabel('Year')
plt.gca().invert_yaxis()
plt.legend(title='Country')
plt.grid(True)
plt.show()

# Trend: GDP per Capita and Social Support
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='GDP per capita', hue='Country or region', data=top_5_countries, marker='o')
plt.title('GDP per Capita Trends for Top 5 Countries (2015-2019)')
plt.ylabel('GDP per Capita')
plt.xlabel('Year')
plt.legend(title='Country')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Social support', hue='Country or region', data=top_5_countries, marker='o')
plt.title('Social Support Trends for Top 5 Countries (2015-2019)')
plt.ylabel('Social Support')
plt.xlabel('Year')
plt.legend(title='Country')
plt.grid(True)
plt.show()
```

```python
bottom_5_countries = data.groupby('Year').apply(lambda x: x.nlargest(5, 'Overall rank')).reset_index(drop=True)

# Trend: Happiness Score
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Score', hue='Country or region', data=bottom_5_countries, marker='o')
plt.title('Happiness Score Trends for Bottom 5 Countries (2015-2019)')
plt.ylabel('Happiness Score')
plt.xlabel('Year')
plt.legend(title='Country')
plt.grid(True)
plt.show()

# Trend: Rank
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Overall rank', hue='Country or region', data=bottom_5_countries, marker='o')
plt.title('Rank Trends for Bottom 5 Countries (2015-2019)')
plt.ylabel('Overall Rank')
plt.xlabel('Year')
plt.gca().invert_yaxis()
plt.legend(title='Country')
plt.grid(True)
plt.show()

# Trend: GDP per Capita and Social Support
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='GDP per capita', hue='Country or region', data=bottom_5_countries, marker='o')
plt.title('GDP per Capita Trends for Bottom 5 Countries (2015-2019)')
plt.ylabel('GDP per Capita')
plt.xlabel('Year')
plt.legend(title='Country')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Social support', hue='Country or region', data=bottom_5_countries, marker='o')
plt.title('Social Support Trends for Bottom 5 Countries (2015-2019)')
plt.ylabel('Social Support')
plt.xlabel('Year')
plt.legend(title='Country')
plt.grid(True)
plt.show()
```
